```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif["Feature"] = df.columns
```

```python
vif['VIF'] =[ variance_inflation_factor(df.values, i) for i in range(len(df.columns))]
```

```python
vif
```

```python
#if vif > 10, drop the features one by one>> here highest is Longitude, we will drop this and then again
#calculate vif and then again drop the highest one

df1 = df.copy()
```

```python
df1
```

```python
df1.drop("Longitude", axis=1, inplace=True)
```

```python
df1
```

```python
vif = pd.DataFrame()
vif["Feature"] = df1.columns
vif['VIF'] =[ variance_inflation_factor(df1.values, i) for i in range(len(df1.columns))]
```

```python
vif
```

```python
df1.drop("AveRooms", axis=1, inplace=True)

vif = pd.DataFrame()
vif["Feature"] = df1.columns
vif['VIF'] =[ variance_inflation_factor(df1.values, i) for i in range(len(df1.columns))]
vif
```

```python
df1.drop("Latitude", axis=1, inplace=True)

vif = pd.DataFrame()
vif["Feature"] = df1.columns
vif['VIF'] =[ variance_inflation_factor(df1.values, i) for i in range(len(df1.columns))]
vif
```

```python
X = df1.iloc[:, :-1] #independent variable
y = df1.iloc[:, -1] #target variable
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
#scaling(optionaal) can be done after train test split

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
#fit_transform is used for train data
#transform is used for test data

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#performance metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
print(r2_score(y_test, y_pred))
```

```python
score = r2_score(y_test, y_pred)
score
```

```python
#internal homework
#adjusted rsquare
1 - ((1-score)*len(y_test)-1) /  ((len(y_test)) - X_test.shape[1]-1)
```
