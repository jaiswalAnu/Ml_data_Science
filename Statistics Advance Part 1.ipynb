**Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution**

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

def bernoulli_pdf(p):
    x = np.array([0, 1])
    return stats.bernoulli.pmf(x, p)

p = 0.7
pdf_values = bernoulli_pdf(p)


plt.stem([0, 1], pdf_values, basefmt=" ", use_line_collection=True)
plt.title(f'Bernoulli Distribution (p = {p})')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.show()
```

**Write a Python function to calculate Z-scores from a dataset and plot them**

```python
import numpy as np
import matplotlib.pyplot as plt
def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores
data = np.random.normal(0, 1, 1000)
z_scores = calculate_z_scores(data)
plt.hist(z_scores, bins=30, density=True, edgecolor='black')
plt.title('Z-scores of Data')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.show()
```

Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribut

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

x, mean, std_dev = 65, 50, 10
z = (x - mean) / std_dev

print(f"Z-score: {z}, Probability: {norm.cdf(z)}")

x_vals = np.linspace(-4, 4, 1000)
plt.plot(x_vals, norm.pdf(x_vals))
plt.axvline(z, color='r', linestyle='--')
plt.fill_between(x_vals, norm.pdf(x_vals), where=(x_vals <= z), color='red', alpha=0.3)
plt.show()
```

**Implement hypothesis testing using Z-statistics for a sample dataset**

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt


sample_data = [52, 48, 50, 55, 49, 53, 51, 54, 47, 50]
pop_mean = 50


z = (np.mean(sample_data) - pop_mean) / (np.std(sample_data, ddof=1) / np.sqrt(len(sample_data)))
p_value = 2 * (1 - norm.cdf(abs(z)))


print(f"Z-stat: {z}, P-value: {p_value}")


x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x))
plt.axvline(z, color='r', linestyle='--')
plt.fill_between(x, norm.pdf(x), where=(abs(x) >= abs(z)), color='red', alpha=0.3)
plt.show()
```

**Create a confidence interval for a dataset using Python and interpret the result**

```python
import numpy as np
import scipy.stats as stats

# Sample dataset
data = [23, 21, 19, 22, 24, 25, 21, 20, 19, 23]

# Calculate sample mean and standard error
mean = np.mean(data)
std_err = np.std(data, ddof=1) / np.sqrt(len(data))

# Confidence level (95%)
confidence_level = 0.95
alpha = 1 - confidence_level

# t-distribution for 95% confidence interval
t_critical = stats.t.ppf(1 - alpha/2, df=len(data)-1)

# Margin of error
margin_of_error = t_critical * std_err

# Confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

**Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean**

```python
import numpy as np
import scipy.stats as stats


np.random.seed(42)
mean_true = 50
std_dev = 10
n = 100


data = np.random.normal(loc=mean_true, scale=std_dev, size=n)


mean_sample = np.mean(data)
std_err = np.std(data, ddof=1) / np.sqrt(len(data))


confidence_level = 0.95
alpha = 1 - confidence_level


t_critical = stats.t.ppf(1 - alpha/2, df=len(data)-1)


margin_of_error = t_critical * std_err


lower_bound = mean_sample - margin_of_error
upper_bound = mean_sample + margin_of_error


print(f"Sample Mean: {mean_sample:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

**Implement Z-score normalization on a dataset and visualize its transformation**

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
mean = 50
std_dev = 10
n = 1000
data = np.random.normal(mean, std_dev, n)
data_mean = np.mean(data)
data_std = np.std(data)
normalized_data = (data - data_mean) / data_std
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='blue', alpha=0.7)
plt.title('Original Data (Before Z-score Normalization)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='orange', alpha=0.7)
plt.title('Normalized Data (After Z-score Normalization)')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):

    data_skewness = skew(data)


    data_kurtosis = kurtosis(data)

    return data_skewness, data_kurtosis
np.random.seed(42)
mean = 50
std_dev = 10
n = 1000
data = np.random.normal(mean, std_dev, n)
skewness, kurt = calculate_skewness_kurtosis(data)
print(f"Skewness of the dataset: {skewness:.2f}")
print(f"Kurtosis of the dataset: {kurt:.2f}")
```
