```python
import requests
import pandas as pd
import time

# Initialize an empty DataFrame
df = pd.DataFrame(columns=['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count'])

# API Key
api_key = 'e29a1a2cfe58721a77ce730876446751'

# Loop through the pages
for i in range(1, 429):
    try:
        response = requests.get(
            f'https://api.themoviedb.org/3/movie/top_rated?api_key={api_key}&language=en-US&page={i}'
        )
        response.raise_for_status()  # Raise an error for bad responses (4xx, 5xx)

        temp_df = pd.DataFrame(response.json()['results'])[
            ['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']
        ]

        df = pd.concat([df, temp_df], ignore_index=True)

        print(f"Page {i} fetched successfully.")

        # Pause to respect API rate limits
        time.sleep(0.5)

    except requests.exceptions.RequestException as e:
        print(f"Error on page {i}: {e}")
        break

# Save the DataFrame to a CSV file
df.to_csv('top_rated_movies.csv', index=False)
print("Data saved to top_rated_movies.csv")
```
