```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Headers to mimic a browser visit
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36",
    "Accept-Language": "en-US, en;q=0.5"
}

def get_amazon_product_data(url):
    response = requests.get(url, headers=HEADERS)
    if response.status_code != 200:
        print("Failed to fetch page, status code:", response.status_code)
        return None

    soup = BeautifulSoup(response.content, "html.parser")

    # Extract Product Title
    title = soup.find("span", {"id": "productTitle"})
    title = title.get_text(strip=True) if title else "N/A"

    # Extract Product Price
    price = soup.find("span", {"class": "a-price-whole"})
    price = price.get_text(strip=True) if price else "N/A"

    # Extract Product Rating
    rating = soup.find("span", {"class": "a-icon-alt"})
    rating = rating.get_text(strip=True) if rating else "N/A"

    # Extract Top Review (if available)
    review = soup.find("span", {"data-hook": "review-body"})
    review = review.get_text(strip=True) if review else "N/A"

    return {
        "Title": title,
        "Price": price,
        "Rating": rating,
        "Top Review": review
    }

# Example Product URL (Replace with your target product URL)
url = "https://www.amazon.com/dp/B08N5WRWNW"
product_data = get_amazon_product_data(url)

# Save to a DataFrame
if product_data:
    df = pd.DataFrame([product_data])
    print(df)
    df.to_csv("amazon_product_data.csv", index=False)
```
