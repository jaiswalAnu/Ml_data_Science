Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results

```python
import numpy as np
import scipy.stats as stats

# Simulate random sample data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)

# Population parameters
population_mean = 50
population_std = 10

# Z-test
sample_mean = np.mean(sample_data)
std_error = population_std / np.sqrt(len(sample_data))
z_stat = (sample_mean - population_mean) / std_error
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Z-statistic: {z_stat}, P-value: {p_value}")
```

Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate sample data (mean = 50, std = 10, sample size = 30)
sample_data = np.random.normal(loc=50, scale=10, size=30)

# Population parameters
population_mean = 50
population_std = 10

# Calculate sample mean and standard error
sample_mean = np.mean(sample_data)
std_error = population_std / np.sqrt(len(sample_data))

# Z-test calculation
z_stat = (sample_mean - population_mean) / std_error
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed test

# Decision region (alpha = 0.05)
z_critical = stats.norm.ppf(1 - 0.025)  # Two-tailed critical value

# Plot Z-distribution and decision region
z_range = np.linspace(-4, 4, 1000)
pdf_values = stats.norm.pdf(z_range)
plt.plot(z_range, pdf_values, label="Z-distribution")
plt.fill_between(z_range, 0, pdf_values, where=(z_range < -z_critical) | (z_range > z_critical), color='red', alpha=0.5)
plt.axvline(z_critical, color='red', linestyle='--')
plt.axvline(-z_critical, color='red', linestyle='--')
plt.title('Two-tailed Z-test Decision Region')
plt.legend()
plt.show()

# Output results
print(f"Z-statistic: {z_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject null hypothesis: Sample mean is not significantly different from population mean.")
```

Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_type_errors(alpha=0.05, effect_size=0.5):
    # Define the null and alternative distributions
    z_range = np.linspace(-4, 4, 1000)
    pdf_null = stats.norm.pdf(z_range, loc=0, scale=1)  # Null hypothesis (H0)
    pdf_alt = stats.norm.pdf(z_range, loc=effect_size, scale=1)  # Alternative hypothesis (H1)

    # Critical values for Type 1 error (alpha) and Type 2 error (beta)
    rejection_region_left = stats.norm.ppf(alpha/2)
    rejection_region_right = stats.norm.ppf(1 - alpha/2)

    # Type 1 error (alpha): False positive, rejecting H0 when it's true
    type_1_error = np.trapz(pdf_null[(z_range < rejection_region_left) | (z_range > rejection_region_right)], z_range[(z_range < rejection_region_left) | (z_range > rejection_region_right)])

    # Type 2 error (beta): False negative, failing to reject H0 when H1 is true
    type_2_error = np.trapz(pdf_alt[(z_range > rejection_region_left) & (z_range < rejection_region_right)], z_range[(z_range > rejection_region_left) & (z_range < rejection_region_right)])

    # Plot
    plt.plot(z_range, pdf_null, label="Null Hypothesis (H₀)")
    plt.plot(z_range, pdf_alt, label="Alternative Hypothesis (H₁)")
    plt.fill_between(z_range, 0, pdf_null, where=(z_range < rejection_region_left) | (z_range > rejection_region_right), color='red', alpha=0.5, label="Type 1 Error (α)")
    plt.fill_between(z_range, 0, pdf_alt, where=(z_range > rejection_region_left) & (z_range < rejection_region_right), color='blue', alpha=0.5, label="Type 2 Error (β)")
    plt.axvline(rejection_region_left, color='red', linestyle='--')
    plt.axvline(rejection_region_right, color='red', linestyle='--')
    plt.title('Type 1 and Type 2 Errors Visualization')
    plt.legend()
    plt.show()

    # Return the error values
    return type_1_error, type_2_error

# Example: Plot Type 1 and Type 2 errors
type_1, type_2 = plot_type_errors(alpha=0.05, effect_size=0.5)
print(f"Type 1 Error (α): {type_1}")
print(f"Type 2 Error (β): {type_2}")
```

Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(sample_data, confidence_level=0.95):
    # Calculate sample mean and standard error
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (using ddof=1 for unbiased estimate)
    sample_size = len(sample_data)
    std_error = sample_std / np.sqrt(sample_size)

    # Calculate the critical value for the given confidence level
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size-1)

    # Calculate margin of error
    margin_of_error = critical_value * std_error

    # Confidence Interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, margin_of_error

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Simulate data: mean=50, std=10, size=30
confidence_level = 0.95

lower, upper, margin = calculate_confidence_interval(sample_data, confidence_level)

print(f"Confidence Interval: ({lower:.2f}, {upper:.2f})")
print(f"Margin of Error: {margin:.2f}")
```

Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayesian_inference(prior, likelihood, evidence):
    """
    Calculate posterior probability using Bayes' Theorem.

    Parameters:
    prior (float): The prior probability P(H)
    likelihood (float): The likelihood P(D|H)
    evidence (float): The evidence P(D)

    Returns:
    float: Posterior probability P(H|D)
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Example: Let's say we want to determine the probability of a person having a disease (H)
# based on a positive test result (D).

# Prior: The probability that a person has the disease (before seeing the test result)
prior = 0.01  # 1% chance of having the disease

# Likelihood: The probability of testing positive given that the person has the disease
likelihood = 0.95  # 95% of people with the disease test positive

# Evidence: The probability of testing positive (regardless of disease)
# This includes people who test positive but do not have the disease (false positives).
evidence = 0.05  # 5% of the population tests positive

# Compute posterior probability using Bayes' Theorem
posterior = bayesian_inference(prior, likelihood, evidence)

print(f"Posterior probability of having the disease given a positive test result: {posterior:.4f}")
```

Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data

```python
import numpy as np

def calculate_expected_frequencies(observed_data):
    # Calculate row totals, column totals, and grand total
    row_totals = np.sum(observed_data, axis=1)
    column_totals = np.sum(observed_data, axis=0)
    grand_total = np.sum(observed_data)

    # Calculate the expected frequencies for each cell
    expected_frequencies = np.outer(row_totals, column_totals) / grand_total

    return expected_frequencies

# Example observed data (contingency table)
# Let's assume we have the following data about 'Gender' and 'Purchased' (Yes/No)
observed_data = np.array([[30, 10],  # 30 males purchased, 10 males didn't purchase
                          [20, 40]]) # 20 females purchased, 40 females didn't purchase

# Calculate expected frequencies
expected_data = calculate_expected_frequencies(observed_data)

# Print the observed and expected frequencies
print("Observed Frequencies:")
print(observed_data)

print("\nExpected Frequencies:")
print(expected_data)
```

16. Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
import scipy.stats as stats

# Step 1: Generate two random samples from normal distributions
np.random.seed(42)  # Set a seed for reproducibility
sample1 = np.random.normal(loc=50, scale=10, size=100)  # Sample 1: mean=50, std=10
sample2 = np.random.normal(loc=50, scale=12, size=100)  # Sample 2: mean=50, std=12

# Step 2: Calculate the sample variances
var1 = np.var(sample1, ddof=1)  # Sample variance of sample1 (ddof=1 for sample variance)
var2 = np.var(sample2, ddof=1)  # Sample variance of sample2 (ddof=1 for sample variance)

# Step 3: Calculate the F-statistic
# F-statistic is the ratio of the larger variance to the smaller variance
F_statistic = var1 / var2 if var1 > var2 else var2 / var1

# Step 4: Degrees of freedom
df1 = len(sample1) - 1  # Degrees of freedom for sample 1
df2 = len(sample2) - 1  # Degrees of freedom for sample 2

# Step 5: Calculate the p-value using the F-distribution
p_value = 1 - stats.f.cdf(F_statistic, df1, df2)  # Upper tail p-value

# Step 6: Output results
print(f"F-statistic: {F_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Interpretation of the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
```

Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVAD

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(groups):
    """
    Check the assumptions of normality, independence, and equal variance for ANOVA.

    Parameters:
    - groups: List of arrays, where each array represents a group of data.

    Returns:
    - None: Prints out the results of the assumption checks.
    """

    # 1. Check Normality: Use Shapiro-Wilk test for normality
    print("Checking Normality (Shapiro-Wilk test):")
    for i, group in enumerate(groups, start=1):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i}: Statistic = {stat:.4f}, p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"Group {i}: Data is not normally distributed (reject null hypothesis).")
        else:
            print(f"Group {i}: Data is normally distributed (fail to reject null hypothesis).")

    # 2. Check Equal Variance: Use Levene's test for homogeneity of variances
    print("\nChecking Equal Variance (Levene's test):")
    stat, p_value = stats.levene(*groups)
    print(f"Levene's test: Statistic = {stat:.4f}, p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("Reject null hypothesis: Variances are not equal (homoscedasticity assumption violated).")
    else:
        print("Fail to reject null hypothesis: Variances are equal (homoscedasticity assumption met).")

    # 3. Independence: Cannot be directly tested, but ensure data collection ensures independence.
    print("\nIndependence Assumption: This assumption cannot be directly tested, but ensure data collection ensures independence.")

    # Visualize data distribution (Normality check visually with Q-Q plot)
    print("\nVisualizing data distribution with Q-Q plots for normality check:")
    plt.figure(figsize=(12, 6))
    for i, group in enumerate(groups, start=1):
        plt.subplot(1, len(groups), i)
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f"Group {i}")
    plt.tight_layout()
    plt.show()

# Example usage
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

check_anova_assumptions([group1, group2, group3])
```

20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
results

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate data
np.random.seed(42)

# Factors (independent variables)
factor1 = ['A', 'B', 'C']
factor2 = ['X', 'Y']

# Simulating the dependent variable (response) data
data = []
for f1 in factor1:
    for f2 in factor2:
        # Random data for each combination of factor levels
        response = np.random.normal(loc=10, scale=5, size=30) + 5 * (factor1.index(f1) - 1) + 3 * (factor2.index(f2) - 1)
        for r in response:
            data.append([f1, f2, r])

# Convert data to a pandas DataFrame
df = pd.DataFrame(data, columns=['Factor1', 'Factor2', 'Response'])

# Step 2: Perform Two-Way ANOVA
model = ols('Response ~ C(Factor1) * C(Factor2)', data=df).fit()  # C() indicates categorical variables
anova_table = sm.stats.anova_lm(model, typ=2)  # Type 2 ANOVA
print(anova_table)

# Step 3: Interpret the results
alpha = 0.05  # Significance level
if anova_table['PR(>F)']['C(Factor1)'] < alpha:
    print("\nFactor1 has a significant effect on the response.")
else:
    print("\nFactor1 does not have a significant effect on the response.")

if anova_table['PR(>F)']['C(Factor2)'] < alpha:
    print("Factor2 has a significant effect on the response.")
else:
    print("Factor2 does not have a significant effect on the response.")

if anova_table['PR(>F)']['C(Factor1):C(Factor2)'] < alpha:
    print("There is a significant interaction between Factor1 and Factor2.")
else:
    print("There is no significant interaction between Factor1 and Factor2.")

# Step 4: Visualize the results
# Plot interaction plot
plt.figure(figsize=(8, 6))
sns.pointplot(data=df, x='Factor1', y='Response', hue='Factor2', dodge=True, markers=["o", "s"], linestyles=["-", "--"])
plt.title("Interaction Plot for Two-Way ANOVA")
plt.xlabel('Factor 1')
plt.ylabel('Response')
plt.legend(title='Factor 2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Function to visualize F-distribution
def visualize_f_distribution():
    # Define degrees of freedom for numerator and denominator
    df1_values = [1, 5, 10]
    df2 = 10

    # Set up the figure for plotting
    plt.figure(figsize=(10, 6))

    # Generate a range of values for the x-axis (F-statistic values)
    x = np.linspace(0, 5, 1000)

    # Plot the F-distribution for different degrees of freedom (df1)
    for df1 in df1_values:
        y = stats.f.pdf(x, df1, df2)  # Probability density function of F-distribution
        plt.plot(x, y, label=f'df1={df1}, df2={df2}')

    # Add labels and title
    plt.title('F-distribution for Different Degrees of Freedom')
    plt.xlabel('F-statistic')
    plt.ylabel('Probability Density')
    plt.legend(title="Degrees of Freedom", loc="upper right")
    plt.grid(True)

    # Show plot
    plt.show()

# Call the function to visualize the F-distribution
visualize_f_distribution()
```

Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate random sample data for three groups
np.random.seed(42)

# Simulating data for three groups
group1 = np.random.normal(loc=20, scale=5, size=30)  # Group 1 (mean=20, std=5)
group2 = np.random.normal(loc=22, scale=5, size=30)  # Group 2 (mean=22, std=5)
group3 = np.random.normal(loc=25, scale=5, size=30)  # Group 3 (mean=25, std=5)

# Step 2: Perform the one-way ANOVA test
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Print the results
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Step 3: Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The group means are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in the group means.")

# Step 4: Visualize the results using a boxplot
# Combine the data into a DataFrame for easy plotting
data = pd.DataFrame({
    'Group1': group1,
    'Group2': group2,
    'Group3': group3
})

# Create a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.title('One-Way ANOVA: Group Comparison')
plt.xlabel('Group')
plt.ylabel('Values')
plt.xticks([0, 1, 2], ['Group 1', 'Group 2', 'Group 3'])
plt.show()
```

Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
import scipy.stats as stats

# Step 1: Simulate some sample data
np.random.seed(42)

# Simulating data (sample size n=30, population variance hypothesized to be 25)
sample_data = np.random.normal(loc=50, scale=5, size=30)
sample_variance = np.var(sample_data, ddof=1)  # Sample variance (ddof=1 for unbiased estimate)
n = len(sample_data)  # Sample size

# Step 2: Hypothesized population variance (e.g., 25)
sigma_squared_0 = 25

# Step 3: Calculate the Chi-square statistic
chi_square_statistic = (n - 1) * sample_variance / sigma_squared_0

# Step 4: Perform the hypothesis test (two-tailed)
# Degrees of freedom = n - 1
df = n - 1
alpha = 0.05

# Step 5: Calculate the p-value
p_value = 2 * min(stats.chi2.cdf(chi_square_statistic, df), 1 - stats.chi2.cdf(chi_square_statistic, df))

# Step 6: Print the results
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-square Statistic: {chi_square_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Interpretation
if p_value < alpha:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized value.")
else:
    print("Fail to reject the null hypothesis: The population variance is not significantly different from the hypothesized value.")
```

Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
import numpy as np
import scipy.stats as stats

# Step 1: Simulate two datasets with binary outcomes (successes/failures)
np.random.seed(42)

# Group 1: 100 successes out of 500 trials
successes_1 = 100
n_1 = 500
# Group 2: 120 successes out of 500 trials
successes_2 = 120
n_2 = 500

# Step 2: Calculate the sample proportions
p_1 = successes_1 / n_1
p_2 = successes_2 / n_2

# Step 3: Calculate the pooled proportion
p_pool = (successes_1 + successes_2) / (n_1 + n_2)

# Step 4: Calculate the Z-statistic
z_statistic = (p_1 - p_2) / np.sqrt(p_pool * (1 - p_pool) * (1 / n_1 + 1 / n_2))

# Step 5: Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed test

# Step 6: Print results
print(f"Sample Proportion 1: {p_1:.4f}")
print(f"Sample Proportion 2: {p_2:.4f}")
print(f"Pooled Proportion: {p_pool:.4f}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")
```

26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate two datasets
np.random.seed(42)

# Sample 1: 100 data points with variance 25
data_1 = np.random.normal(loc=50, scale=5, size=100)

# Sample 2: 100 data points with variance 50
data_2 = np.random.normal(loc=50, scale=7, size=100)

# Step 2: Calculate the sample variances
var_1 = np.var(data_1, ddof=1)  # ddof=1 for sample variance
var_2 = np.var(data_2, ddof=1)

# Step 3: Calculate the F-statistic (larger variance in the numerator)
f_statistic = var_1 / var_2 if var_1 > var_2 else var_2 / var_1

# Step 4: Calculate degrees of freedom for both samples
df1 = len(data_1) - 1  # degrees of freedom for sample 1
df2 = len(data_2) - 1  # degrees of freedom for sample 2

# Step 5: Calculate the p-value using the F-distribution
p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

# Step 6: Print results
print(f"Sample Variance 1: {var_1:.4f}")
print(f"Sample Variance 2: {var_2:.4f}")
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Visualization of F-distribution
x = np.linspace(0, 5, 500)
y = stats.f.pdf(x, df1, df2)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})', color='b')
plt.fill_between(x, y, where=(x >= f_statistic), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(f_statistic, color='r', linestyle='--', label=f'F-statistic: {f_statistic:.4f}')
plt.title('F-distribution and F-test')
plt.xlabel('F-value')
plt.ylabel('Density')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()

# Step 8: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")

```

Perform a Chi-square test for goodness of fit with simulated data and analyze the results

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate the rolling of a fair six-sided die 500 times
np.random.seed(42)
observed_data = np.random.choice([1, 2, 3, 4, 5, 6], size=500)

# Step 2: Calculate observed frequencies
observed_freq = np.array([np.sum(observed_data == i) for i in range(1, 7)])

# Step 3: Define expected frequencies for a fair die (equal for each side)
expected_freq = np.array([500 / 6] * 6)

# Step 4: Perform the Chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(observed_freq, expected_freq)

# Step 5: Print results
print(f"Observed Frequencies: {observed_freq}")
print(f"Expected Frequencies: {expected_freq}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 6: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The die is not fair.")
else:
    print("Fail to reject the null hypothesis: The die appears to be fair.")

# Step 7: Visualization of observed vs expected frequencies
categories = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(categories))

plt.bar(x - 0.2, observed_freq, 0.4, label='Observed', color='blue')
plt.bar(x + 0.2, expected_freq, 0.4, label='Expected', color='red')

plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Chi-square Goodness of Fit Test: Observed vs Expected Frequencies')
plt.xticks(x, categories)
plt.legend()
plt.show()
```
