```python
import pandas as pd
from itertools import product
```

```python
animal = ['cat', 'dog', 'cow', 'elephant']
age = [20, 30, 40, 50, 60]
color = ['brown', 'black', 'white']
habitat = ['domestic', 'jungle', 'water']
```

```python
combinations = list(product(animal, age, color, habitat))
```

```python
df = pd.DataFrame(combinations, columns=['animal', 'age', 'color', 'habitat'])
```

```python
df.to_csv('combinations.csv', index=False)
```

```python
num_rows = len(df)
```

```python
print("Combination file has", num_rows, "rows.")
```

```python
def generate_dict(max_num=15):
    result_dict = {}

    for num in range(1, max_num + 1):
        square = num ** 2
        cube = num ** 3

        if square % 3 != 0:
            result_dict[num] = (square, cube)

    return result_dict
result = generate_dict()
print(result)
```

```python
import datetime
data = {
    'name': 'John Doe',
    'age': 30,
    'contact': {
        'phone': '555-1234',
        'email': 'john.doe@example.com'
    },
    'orders': [
        {
            'id': 1,
            'purchase_date': datetime.datetime(2020, 4, 27, 13, 30),
            'items': [
                {'product': 'Widget A', 'quantity': 2, 'transaction_charge': 9.99},
                {'product': 'Widget B', 'quantity': 1, 'transaction_charge': 12.99}
            ]
        },
        {
            'id': 2,
            'purchase_date': datetime.datetime(2022, 4, 26, 13, 30),
            'items': [
                {'product': 'Widget A', 'quantity': 3, 'transaction_charge': 9.99},
                {'product': 'Widget C', 'quantity': 1, 'transaction_charge': 19.99}
            ]
        }
    ]
}
first_order_date = data['orders'][0]['purchase_date']
last_order_date = data['orders'][-1]['purchase_date']
time_difference = last_order_date - first_order_date
months = time_difference.days // 30
days = time_difference.days % 30
print("Time difference between the first and last order:", months, "months and", days, "days")
product_transaction_charges = {}
for order in data['orders']:
    for item in order['items']:
        product = item['product']
        transaction_charge = item['transaction_charge']
        if product in product_transaction_charges:
            product_transaction_charges[product].append(transaction_charge)
        else:
            product_transaction_charges[product] = [transaction_charge]
average_transaction_charges = {}
for product, charges in product_transaction_charges.items():
    average_charge = sum(charges) / len(charges)
    average_transaction_charges[product] = average_charge
print("Average transaction charge for each product type:", average_transaction_charges)
```
