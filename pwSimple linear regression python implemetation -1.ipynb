```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the data
#here we are loading the data from scikit learn
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes
```

```python
#Problem statement: To predict disease progression one year after baseline provided 10 features
#in one year how the sugar level has changed
```

```python
data = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
```

```python
data.head()
```

```python
data['target'] = diabetes.target
```

```python
data
```

```python
#EDA, DATACLEANING, DATAPREPARATION, FEATURE ENGINEERING
```

```python
X = data[["bmi"]] #one indepenedent variable
y = data['target'] #dependent variable
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=2)
#reandom_state when fixed helps us to reproduce the same result again and again
```

```python
X_train.shape
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
#scaling(optional)
```

```python
#model training
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
X_train
```

```python
y_train
```

```python
#training of model
model.fit(X_train, y_train)
#at the backend it will use gradient descent to give optimal coefficient
```

```python
model.coef_ #attribute of the model
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#visualise the result
plt.scatter(X_test, y_test, color = 'black', label = "Actual")
plt.plot(X_test, y_pred, color = 'blue', linewidth = 3, label = "Predicted linear regression line")
plt.xlabel("BMI")
plt.ylabel("One year progression target")
plt.title("Linear regression on diabetes data")
plt.legend()
plt.show()
```

```python
#model evaluation
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```
