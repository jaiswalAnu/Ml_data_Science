```python
n=int(input("Enter the number"))
def generate_string(n):
  result=[]
  for i in range(1,n+1):
      string=str(i)
      if i%4==0 and i%6==0:
        string="QuadHex"
      elif i%4==0:
        string="Quad"
      elif i%6==0:
        string="Hex"
      result.append(string)
  return result
output=generate_string(n)
print(output)
```

```python
def is_valid_vehicle_number(s):
  for i in range(27):
    substring = s[i:i+10]
    if (
            len(substring) == 10
            and substring[0:2].isalpha()
            and substring[2:4].isdigit()
            and substring[4:6].isalpha()
            and substring[6:].isdigit()
        ):
      return True
    return False


s1 = "AB12MH14AB1234YZ78GH9012JKL34MN56"
s2 = "XX11YY2A33ZZ44AA55X6BB77881C99"

print(is_valid_vehicle_number(s1))
print(is_valid_vehicle_number(s2))
```

```python
import re

def is_valid_vehicle_number(s):
    # Define the pattern for a valid Indian vehicle number
    pattern = r"[A-Z]{2}[0-9]{2}[A-Z]{2}[0-9]{4}"

    # Search for the pattern in the string
    match = re.search(pattern, s)

    # Return True if a match is found, otherwise False
    return bool(match)

# Test cases
test_cases = [
    "AB12MH14AB1234YZ78GH9012JKL34MN56",  # Expected Output: True
    "XX11YY2A33ZZ44AA55X6BB77881C99",     # Expected Output: False
    "AB12CD34EF5678GH90IJ23KL45MN6789",   # Expected Output: True
    "AA00BB0001CC02DD03EE04FF05GG06HH07", # Expected Output: True
    "ZZ11YY22XX33AA44BB55CC66DD77EE88"    # Expected Output: False
]

# Running test cases
for idx, test in enumerate(test_cases, 1):
    result = is_valid_vehicle_number(test)
    print(f"Test case {idx}: {result}")
```
